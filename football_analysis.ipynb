```python

with open('Youth_Player_Development_Data.csv', 'r', encoding='latin1') as f:
    lines = f.readlines()
    players = []

    for line in lines:
        data= line.strip().strip(',')
        player = data.strip().split(',')
        players.append(player)


print(players[1])

def nations(nationality, list):

    for i in players:
        if players[i][1] == nationality:
            list.append(players[i])
    return list
```

```python
def nations(nationality, list):

    for i in range(len(players)):
        if players[i][1] == nationality:
            list.append(players[i])
    return list

hungarian_players = []
nations('Hungarian', hungarian_players)
hungarian_players[-1]

croatian_players = []
nations('Croatian', croatian_players)
croatian_players[-1]

austrian_players = []
nations('Austrian', austrian_players)
austrian_players[-1]

czech_palyers = []
nations('Czech', czech_palyers)

serbian_players = []
nations('Serbian', serbian_players)
```

```python
import matplotlib.pyplot as plt
hungarian_young_mins = []
for i in range(len(hungarian_players)):
    hungarian_young_mins.append(hungarian_players[i][2])
hungarian_MVs = []
for i in range(len(hungarian_players)):
    hungarian_MVs.append(hungarian_players[i][4])



croatian_young_mins = []
for i in range(len(croatian_players)):
    croatian_young_mins.append(croatian_players[i][2])
croatian_MVs = []
for i in range(len(croatian_players)):
    croatian_MVs.append(croatian_players[i][4])


austrian_young_mins = []
for i in range(len(austrian_players)):
    austrian_young_mins.append(austrian_players[i][2])
austrian_MVs = []
for i in range(len(austrian_players)):
    austrian_MVs.append(austrian_players[i][4])

czech_young_mins = []
for i in range(len(czech_palyers)):
    czech_young_mins.append(czech_palyers[i][2])
czech_MVs = []
for i in range(len(czech_palyers)):
    czech_MVs.append(czech_palyers[i][4])

serbian_young_mins = []
for i in range(len(serbian_players)):
    serbian_young_mins.append(serbian_players[i][2])
serbian_MVs = []
for i in range(len(serbian_players)):
    serbian_MVs.append(serbian_players[i][4])

player_data = []

for i in range(len(hungarian_MVs)):
    player_data.append((hungarian_young_mins[i], hungarian_MVs[i], 'hungarian'))

for i in range(len(croatian_MVs)):
    player_data.append((croatian_young_mins[i], croatian_MVs[i], 'croatian'))

for i in range(len(austrian_MVs)):
    player_data.append((austrian_young_mins[i], austrian_MVs[i], 'austrian'))

for i in range(len(czech_MVs)):
    player_data.append((czech_young_mins[i], czech_MVs[i], 'czech'))

for i in range(len(serbian_MVs)):
    player_data.append((serbian_young_mins[i], serbian_MVs[i], 'serb'))

    
```

```python
player_data.sort(key = lambda x: x[0])
young_mins = [player[0] for player in player_data]
MVs = [player[1] for player in player_data]
nationalities = [player[2] for player in player_data]
print(player_data)
```

```python
colors = {'hungarian': 'green', 'austrian': 'red', 'croatian' : 'blue', 'czech': 'purple', 'serb': 'yellow'}
plt.figure(figsize=(6, 5))
plotted_labels = []

for i in range(len(player_data)):

    country  = player_data[i][2]  

    
    
    plt.scatter(int(young_mins[i]), float(MVs[i]),     label = country if country not in plotted_labels else None
,color = colors[country])
    plotted_labels.append(country)



plt.title('How do senior minutes influence market value at big transfer?')
plt.xlabel('Senior minutes (age: 16-20)', size = 12)
plt.ylabel('Market value at big transfer', size = 12)
plt.xticks()
plt.ticklabel_format(style='plain', axis='y')

plt.yticks()

plt.legend()


```
